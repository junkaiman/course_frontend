```python
from bs4 import BeautifulSoup

def get_profs(filename):
    instructor_names = set()
    data = open(filename).readlines()
    # win0divMTG_INSTR
    data = filter(lambda x: 'win0divMTG_INSTR' in x, data)
    data = list(data)
    for element in data:
        # Use BeautifulSoup to parse the HTML element
        soup = BeautifulSoup(element, 'html.parser')    
        instructor_names.add(soup.span.text.strip().replace(',', ''))
    
    return instructor_names
```

```python
sp19 = get_profs('2019spring.txt')
fa19 = get_profs('2019fall.txt')
sp20 = get_profs('2020spring.txt')
fa20 = get_profs('2020fall.txt')
sp21 = get_profs('2021spring.txt')
fa21 = get_profs('2021fall.txt')
sp22 = get_profs('2022spring.txt')
fa22 = get_profs('2022fall.txt')
sp23 = get_profs('2023spring.txt')
```

```python
total = sp19 | fa19 | sp20 | fa20 | sp21 | fa21 | sp22 | fa22 | sp23
total = list(total)
total.sort()
total
```

```python
# to json, utf-8
import json
with open('profs.json', 'w', encoding='utf-8') as f:
    json.dump(total, f, ensure_ascii=False)
```
