```python
import csv
import hashlib
import json

# Initialize an empty list to store the data
data = []

# Open the CSV file and read the contents
with open('dku_course_list_20230104.csv', 'r') as file:
  reader = csv.reader(file)
  
  # Initialize a counter to keep track of the total number of items
  counter = 1000
  
  # Iterate over the rows in the CSV file
  for row in reader:
    # Split the row into two parts: the course code and the course name
    code, name = row
    
    # Extract the department from the course code
    department = code.split(' ')[0]
    
    # Check if the department already exists in the data list
    department_data = next((item for item in data if item["label"] == department), None)
    
    # If the department does not exist, create a new entry for it
    if department_data is None:
      department_data = {"label": department, "items": []}
      data.append(department_data)
    
    # Add the course to the list of items for the department
    # department_data["items"].append({"label": f"{code} - {name}", "to": f"/view/{counter}"})
    hash_code = hashlib.sha1(code.encode('utf-8')).hexdigest()[:10]
    department_data["items"].append({"label": f"{code} - {name}", "to": f"/view/{hash_code}"})
    
    # Increment the counter
    counter += 1

# Convert the data list to a JSON object and save it to a file
with open('output.json', 'w') as file:
  json.dump({"root": data}, file)
```
